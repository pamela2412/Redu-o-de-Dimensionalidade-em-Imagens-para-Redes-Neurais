```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image  #

# Cargar la imagen
input_image = mpimg.imread('cachorro.png')




plt.show()
def escala_grises(input_image):
    # Separar los canales RGB
    red_channel = input_image[:, :, 0]
    green_channel = input_image[:, :, 1]
    blue_channel = input_image[:, :, 2]
    # Valor de Gamma y constantes
    gamma = 1.04
    red_constant = 0.2989
    green_constant = 0.5870
    blue_constant = 0.1140
    # Frmula para conversin a escala de grises
    grayscale_image = (red_constant * red_channel ** gamma) + (green_constant * green_channel ** gamma) + (blue_constant * blue_channel ** gamma)
    fig = plt.figure(figsize=(12, 6))
    # Subgrfico para imagen original
    img1 = fig.add_subplot(121)
    img1.imshow(input_image)
    img1.set_title("Imagen Original")
    img1.axis('off')  # Desactivar los ejes

    # Subgrfico para imagen en escala de grises
    img2 = fig.add_subplot(122)
    img2.imshow(grayscale_image, cmap='gray')
    img2.set_title("Imagen en Escala de Grises")
    img2.axis('off')  # Desactivar los ejes
    plt.show()
# Funcin para convertir a blanco y negro
def blanco_negro(inputimage, grisBase):
    imagen=input_image.copy()
    # Convertir la imagen de Numpy a un objeto PIL
    im = Image.fromarray((imagen * 255).astype(np.uint8))  # Convertimos a uint8 (0-255)
    im = im.convert('RGB')  # Asegurarse de que la imagen est en formato RGB

    im7 = im.copy()  # Copiar la imagen para no modificar la original
    width, height = im7.size  # Obtener las dimensiones de la imagen

    # Binarizacin de la imagen
    for i in range(width):
        for j in range(height):
            r, g, b = im7.getpixel((i, j))
            gris = (r + g + b) / 3
            if gris < grisBase:
                im7.putpixel((i, j), (0, 0, 0))  # Negro
            else:
                im7.putpixel((i, j), (255, 255, 255))  # Blanco
    fig = plt.figure(figsize=(12, 6))

    # Subgrfico para imagen original
    img1 = fig.add_subplot(121)
    img1.imshow(input_image)
    img1.set_title("Imagen Original")
    img1.axis('off')  # Desactivar los ejes

    # Subgrfico para imagen en blanco y negro
    img2 = fig.add_subplot(122)
    img2.imshow(im7)
    img2.set_title("Imagen Blanco y Negro")
    img2.axis('off')  # Desactivar los ejes

    plt.show()


# Llamar a la funcin con un umbral para la binarizacin

blanco_negro(input_image, 128)  # 128 es un valor comn para el umbral
escala_grises(input_image)



```
